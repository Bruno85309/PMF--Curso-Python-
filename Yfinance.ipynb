### Obtendo dados de cotação

```python
itub.history(start="2023-01-01", end="2023-10-01")
```

```python
# Intervalo acaba dependendo do range das datas (ex: 1 minuto tem ser apenas de 7 dias atrás. 5 minutos tem que ser de 60 dias)

itub.history(start="2023-10-02", end="2023-10-09", interval="5m")
```

### Extração multiticker

```python
lista = ["ITUB3.SA", "VALE3.SA", "PETR4.SA"]

yf.download(lista, start="2023-01-01", end="2023-10-01")
```

* Filtrando por ticker

```python
yf.download(lista, start="2023-01-01", end="2023-10-01", group_by="ticker")
```
