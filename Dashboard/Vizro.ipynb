```python
import fundamentus
import pandas as pd
import matplotlib.pyplot as plt
import vizro.plotly.express as px
from vizro import Vizro
import vizro.models as vm
import yfinance as yf
```

```python
df = fundamentus.get_resultado()
```

```python
df
```

```python
setores = pd.read_excel('SETORESB3.xlsx')
```

```python
setores
```

```python
# Arrumando df para fazer merge com a setores

df['TICKER'] = df.index.str[0:4]
```

```python
df['FINAL'] = df.index.str[4:6]
```

```python
df.head()
```

```python
# Fazendo Merge
resultados = pd.merge(df, setores, how = 'inner', on = 'TICKER')
```

```python
resultados.head()
```

* Bom analisar os dados

```python
resultados.dy.sort_values()
```

```python
# Vamos eliminar as últimas 2 linhas

resultados = resultados.sort_values('dy')[:-2]
```

```python
# Vamos elimina um ROE estranho
resultados = resultados.sort_values('roe')[:-1]
```

```python
resultados
```

## 3. Fazendo Top 10 DY do Brasil 

```python
topdy = resultados.sort_values('dy', ascending= False).head(10)
```

```python
topdy
```

```python
page = vm.Page(
    title="Dashboard dos Maiores DY",
    components=[
        vm.Graph(id="scatter_chart", figure=px.bar(topdy, x="TICKER", y="dy", color="SETOR ECONÔMICO"))
    ],
)

dashboard = vm.Dashboard(pages=[page])

Vizro().build(dashboard).run()
```

```python
page = vm.Page(
    title="Dashboard Fundamentalista",
    components=[
        vm.Graph(id="scatter_pl_pvp", figure=px.scatter(resultados, x="TICKER", y="pvp", color="SETOR ECONÔMICO")),
        vm.Graph(id="bar_roe", figure=px.bar(resultados, x="TICKER", y="roe", color="SETOR ECONÔMICO")),
        vm.Graph(id="bar_dy", figure=px.bar(resultados, x="TICKER", y="dy", color="SETOR ECONÔMICO"))
    ],
    controls = [
        vm.Filter(column= 'SEGMENTO'),
        vm.Filter(column= 'pl', selector= vm.RangeSlider()),
        vm.Filter(column= 'pl', selector= vm.RangeSlider()),
        vm.Filter(column= 'pl', selector= vm.RangeSlider())
    ]
)

dashboard = vm.Dashboard(pages=[page])

Vizro().build(dashboard).run()
```
