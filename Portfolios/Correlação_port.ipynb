```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
carteira = ['PRIO3.SA', 'MILS3.SA', 'GGPS3.SA', 'VAMO3.SA', 'TTEN3.SA','VBBR3.SA', 'SBSP3.SA', 'CYRE3.SA', 'INBR32.SA']
```

```python
len(carteira)
```

```python
dados = yf.download(carteira, start='2023-01-01')['Adj Close']
```

```python
dados.dropna(inplace=True)
```

```python
dados
```

```python
retornos = dados.pct_change()
```

```python
retornos = retornos.dropna()
retornos
```

```python
correlacao = retornos.corr()
```

```python
correlacao
```

```python
plt.figure(figsize = (16,8))

sns.heatmap(correlacao.round(decimals= 2),
            vmin = -1, vmax = 1, annot = True, cmap = 'BrBG');
```

# PL

```python
primeiro = dados.iloc[0]
```

```python
num_papeis = 1000/primeiro
```

```python
comprados = round(num_papeis, 0)
comprados
```

```python
pl = comprados*dados
pl.head()
```

```python
pl['PL Total'] = pl.sum(axis = 1)
pl.head()
```

```python
pl['PL Total'].plot();
```
