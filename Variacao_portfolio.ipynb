# Portfólio variando em 2023

```python
# 0 - Importar Bibliotecas

import yfinance as yf
import matplotlib.pyplot
```

```python
# 1 - Selecionar um portfolio

portfolio = ["CVCB3", "ITUB4", "RENT3", "PETZ3", "PETR4", "HAPV3", "MGLU3", "RRRP3", "BHIA3", "MRVE3"]
```

```python
# 2 - Colocar .SA para puxar os dados do Yahoo Finance

portfolio_02 = []

for i in portfolio:
    portfolio_02.append(i + ".SA")

print(portfolio_02)
```

```python
# 3 - Selecionar uma data de início (e de fim, se quiser) e fazer o download

inicio = "2023-01-01"

precos = yf.download(portfolio_02, inicio)["Adj Close"]
```

```python
# 4 - Normalizar para comparação ficar justa

normal = precos/precos.iloc[0]

normal.iloc[[0, -1]].transpose()*100
```

```python
# 5 - Plotar o gráfico

normal.plot(figsize = (18, 10));
```

```python
normal.to_excel('Dados_Normalizados.xlsx')
```

## Correlação

```python
tabela = yf.download(portfolio_02, start= '2014-01-01')["Adj Close"]
```

```python
retorno = tabela.pct_change()
```

```python
retorno
```

```python
retorno = retorno.dropna()
```

```python
retorno
```

```python
correlacao = retorno.corr()
```

```python
correlacao
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (16,8))

sns.heatmap(correlacao.round(decimals= 2),
            vmin = -1, vmax = 1, annot = True, cmap = 'BrBG');
```
