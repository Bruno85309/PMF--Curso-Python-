## TXT

```python
import pandas as pd
```

```python
# Usar read_csv para ler txt

petr4_txt = pd.read_csv("PETR4.txt")
```

```python
petr4_txt
```

```python
petr4_txt.info()
```

* Observe que a coluna de Date não está no formato "datetime64", mas sim no formato "object"

```python
# Conferindo o formato da coluna

petr4_txt["Date"]
```

```python
# Transformando a coluna Date

petr4_txt["Date"] = pd.to_datetime(petr4_txt["Date"])

petr4_txt["Date"]
```

Como transformamos a coluna 'Date' no index do dataframe?

```python
petr4_txt.index = petr4_txt['Date']
```

```python
petr4_txt.head()
```

```python
# Plotando o gráfico com as mudanças de Date (para datetime) e Date no index

petr4_txt["Close"].plot();
```

```python
# Vendo as colunas

petr4_csv.columns
```

```python
# Vendo o shape

petr4_txt.shape
```

```python
# Vendo os tipos

petr4_txt.dtypes
```

```python
# Obtendo valores e estatísticas. Máx, Mín, Standard-deviation e Média

print(petr4_txt.Close.max())
print(petr4_txt.Close.min())
print(petr4_txt.Close.std())
print(petr4_txt.Close.mean())
```
